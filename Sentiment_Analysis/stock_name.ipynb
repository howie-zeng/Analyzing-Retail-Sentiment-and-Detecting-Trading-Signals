```python
import os
import re
import numpy as np
import pandas as pd
```

```python
file_path = 'data/stock_data.csv'
df = pd.read_csv(file_path)
df[:10]
```

```python
def extract_stock_names(text):
    # all uppercase letters and at least 2 characters long
    stock_names = re.findall(r'\b[A-Z]{2,}\b', text)
    return stock_names

df['StockNames'] = df['Text'].apply(extract_stock_names)

stock_count = {}
for stock_names in df['StockNames']:
    for stock in stock_names:
        if stock in stock_count:
            stock_count[stock] += 1
        else:
            stock_count[stock] = 1
sorted_stock_count = dict(sorted(stock_count.items(), key=lambda item: item[1], reverse=True))
top_50_stocks = list(sorted_stock_count.items())[:50]
```
